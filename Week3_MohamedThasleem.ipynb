```python
# Read data from 
dt = pd.read_csv("https://raw.githubusercontent.com/mohamedthasleem/DATA620/master/Wiki-Vote_snap.txt", sep = "\t", names = ['source','target'],skiprows=4, nrows=300)
```

```python
#Sample Records
print(dt.head())
```

```python
#Data Stats
print(dt.describe())
```

```python
#Data Shape
print(dt.shape)
```

```python
#Returns a graph from Pandas DataFrame containing an edge list.
G = nx.from_pandas_edgelist(dt)
```

```python
#Graph Information
print(nx.info(G))
```

```python
#Diameter
print(nx.algorithms.diameter(G))
```

```python
#Checking the graph is connected
print(nx.is_connected(G))
```

```python
%matplotlib inline
nx.draw(G, node_color='#beebe9', node_shape='o', with_labels='True')
```
